## $\pi^0$ check

````python
def plot_pi0_mass(data):
    plt.hist(data.pi0_M, range = [0.11, 0.16], label = 'All $\pi^0$')
    # plt.hist(loose[(loose.gamma1_E > 0.1) & (loose.gamma2_E > 0.1)].pi0_M, range = [0.11, 0.16], label = 'gamma')
    plt.hist(data[data.pi0_isSignal == 1].pi0_M, range = [0.11, 0.16], label = 'Matched $\pi^0$')
    plt.axvline(x = pdg.get('pi0').Mass(), color = 'red')
    plt.legend()
    plt.title('$\pi^0$ mass distribution')
    plt.xlabel('$\pi^0$ mass (GeV)')
    plt.show()
plot_pi0_mass(data)
````

````python
def plot_pi0_momentum(data):
    plt.hist(data.pi0_p, range = [0, 1], label = 'All $\pi^0$')
    plt.hist(data[data.pi0_isSignal == 1].pi0_p, range = [0, 1], label = 'Matched $\pi^0$')
    plt.legend()
    plt.title('$\pi^0$ momentum distribution')
    plt.xlabel('$\pi^0$ momentum (GeV)')
    plt.show()
plot_pi0_momentum(data)
````

## $\Sigma^+$ check

````python
def plot_sigma(loose):
    mass_range = [1.17, 1.21]
    plt.figure(figsize=(16, 6))
    plt.subplot(1, 2, 1)
    plt.hist(loose.M, range = mass_range, label = '$\Sigma+$ all')
    plt.hist(loose[loose.cosa > 0].M, range = mass_range, label = '$\Sigma+$ cosa > 0')
    plt.hist(loose[loose.pi0_isSignal == 1].M, range = mass_range, label = '$\pi^0$ matched')
    plt.hist(loose[loose.isSignal == 1].M, range = mass_range, label = '$\Sigma+$ matched')
    plt.axvline(x = pdg.get('Sigma+').Mass(), color = 'red')
    plt.legend()
    plt.xlabel('$\Sigma^+$ mass (GeV)')
    plt.title('$\Sigma^+$ mass distribution after some loose cuts')
    plt.subplot(1, 2, 2)
    plt.hist(loose[loose.isSignal == 1].M, range = mass_range, label = '$\Sigma+$ matched')
    plt.axvline(x = pdg.get('Sigma+').Mass(), color = 'red')
    plt.legend()
    plt.xlabel('$\Sigma^+$ mass (GeV)')
    plt.title('$\Sigma^+$ mass distribution after some loose cuts')
    plt.show()

plot_sigma(data)
````
